```python

from scipy import ndimage

def zoom_square_img(img, target_size=32.):
    # Original array: img
    # compute the ratio to pobtain a 32 px imgage
    source_size = img.shape[0]
    ratio = target_size/source_size
    #print((ratio,img.shape))
        
    # Resampled in a image of 32 pixels with nearest interpolation:
    if len(img.shape)>2:
        new_img = ndimage.interpolation.zoom(img, (ratio,ratio,1), order=0)
    else:
        new_img = ndimage.interpolation.zoom(img, ratio, order=0)
    return new_img

def get_size(desired_size,r,full_img):
    x,y = full_img.shape
    xc = np.round(r.centroid[0])
    yc = np.round(r.centroid[1])
    if xc-desired_size < 0:
        desired_size = xc
    if yc-desired_size < 0:
        desired_size = yc
    if xc+desired_size >= x:
        desired_size = x-xc-1
    if yc+desired_size >= y:
        desired_size = y-yc-1
    return desired_size
        

def square_norm_img(r,full_img):
    desired_size = np.round(r.major_axis_length/2)
    xc = np.round(r.centroid[0])
    yc = np.round(r.centroid[1])
    sq_img = []
    # make sure that the image is square (not on the border)
    if len(full_img.shape) > 2:
        desired_size = get_size(desired_size,r,full_img[:,:,0])
        sq_img = full_img[xc-desired_size:xc+desired_size,yc-desired_size:yc+desired_size,:]
    else:   
        desired_size = get_size(desired_size,r,full_img)
        sq_img = full_img[xc-desired_size:xc+desired_size,yc-desired_size:yc+desired_size]

    return zoom_square_img(sq_img)

    
x = np.arange(40*40).reshape(40,40)
#show(x)
x2 = zoom_square_img(x)
#show(x2)
x2.shape
print(x2.flatten().shape)


#im = manip.load_image(idx).mean(2)
#show(im)
#d=regions[0]
#show(bw[400:480,100:300])
#show(im[d.bbox[0]:d.bbox[2],d.bbox[1]:d.bbox[3]])
#x2 = square_norm_img(d,im)
#show(x2)
```

```python
import glob, os
from skimage import io
path_tmp = os.path.abspath("./classif/auto") + "/"
list_auto = glob.glob(os.path.join(path_tmp, '*'))

def trans_in_features(f_img):
    features = f_img.mean(2).flatten()
    #features = zoom_square_img(f_img).flatten()
    return features

x_all = []
y_all = []
k=0
for file_name in list_auto:
    im = io.imread(file_name)
    new_data = trans_in_features(im)
    #print(new_data.shape)
    if k==0:
        x_all = new_data
        y_all = np.array(0)
        k=1
    else:
        x_all = np.vstack((x_all , new_data))
        y_all = np.vstack((y_all , 0))
        
path_tmp = os.path.abspath("./classif/pedestrian") + "/"
list_auto = glob.glob(os.path.join(path_tmp, '*'))
for file_name in list_auto:
    im = io.imread(file_name)
    new_data = trans_in_features(im)
    #print(new_data.shape)
    if k==0:
        x_all = new_data
        y_all = np.array(1)
        k=1
    else:
        x_all = np.vstack((x_all , new_data))
        y_all = np.vstack((y_all , 1))
```

```python
from sklearn.grid_search import GridSearchCV
from sklearn.svm import SVC

##############################################################################
# Train classifier
#
# For an initial search, a logarithmic grid with basis
# 10 is often helpful. Using a basis of 2, a finer
# tuning can be achieved but at a much higher cost.

C_range = 10.0 ** np.arange(-5,2)
gamma_range = 10.0 ** np.arange(-5, 3)
param_grid = dict(gamma=gamma_range, C=C_range)
cv = StratifiedKFold(y=y, n_folds=10)
grid = GridSearchCV(SVC(kernel='rbf', class_weight='auto'), param_grid=param_grid, cv=cv)
grid.fit(x_scaled, encoder.transform(y))

print("The best classifier is: ", grid.best_estimator_)
param_grid
```
